```
print(new_data.head(5))
new_data.drop(['Attrition','OverTime','BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus'],axis=1,inplace=True)
# Since we have already created dummy variables so we can drop the columns with categorical features.
print(new_data.head(5))
```

```
#to convert the categorical data with numerical values
print(data1.columns)
print(data1['BusinessTravel'].unique())
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
data1['BusinessTravel']=label_encoder.fit_transform(data1['BusinessTravel'])
data1['Department']=label_encoder.fit_transform(data1['Department'])
data1['Education']=label_encoder.fit_transform(data1['Education'])
data1['Gender']=label_encoder.fit_transform(data1['Gender'])
data1['JobRole']=label_encoder.fit_transform(data1['JobRole'])
data1['MaritalStatus']=label_encoder.fit_transform(data1['MaritalStatus'])
data1['EducationField']=label_encoder.fit_transform(data1['EducationField'])
data1['Attrition']=label_encoder.fit_transform(data1['Attrition'])
data1['Over18']=label_encoder.fit_transform(data1['Over18'])
data1.drop(['Attrition'],axis=1,inplace=True)
```

```
# To spilit the training and testing data

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data1,target,test_size=0.33,random_state=7)
```

```
# Random_forest
from sklearn.tree import export_graphviz
from sklearn.ensemble import RandomForestClassifier
print('RANDOM FOREST')
forest=RandomForestClassifier(n_estimators=1000)
forest.fit(x_train,y_train.values.ravel())
estimator = forest.estimators_[5]
predicted= forest.predict(x_test)
print(predicted)
```

```
#logistic regression
# import the class
print('LOGISTIC REGRESSION')
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
# fit the model with data
logreg.fit(x_train,y_train.values.ravel())
y_pred=logreg.predict(x_test)
print(y_pred)
```

```

#DECISION TREE CLASSIFICATION
# Create Decision Tree classifer object
print("DECISION TREEE")
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
clf = DecisionTreeClassifier(class_weight=None, criterion='gini', max_depth=None,
            max_features=None, max_leaf_nodes=None, min_samples_leaf=1,
            min_samples_split=20, min_weight_fraction_leaf=0.0,
            presort=False, random_state=None, splitter='best')
# Train Decision Tree Classifer
clf = clf.fit(x_train,y_train.values.ravel())
y_pred_dt = clf.predict(x_test)
print(y_pred_dt)
```

```
#Decision tree model visualisation
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,
filled=True, rounded=True,
special_characters=True,feature_names = data1.columns,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('decision_tree.png')
Image(graph.create_png())
```

```
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
print("Accuracy score (validation): {0:.3f}".format(forest.score(x_test,y_test)))
print("Confusion Matrix for Random Forests:")
print(confusion_matrix(y_test, predicted))
print()
print("Classification Report for Random Forests")
print(classification_report(y_test, predicted))
print("LOGISTIC REGRESSION")
print("Accuracy score (validation): {0:.3f}".format(logreg.score(x_test,y_test)))
print("Confusion Matrix Logistic Regression:")
print(confusion_matrix(y_test, y_pred))
print()
print("Classification Report for Logistic Regression")
print(classification_report(y_test, y_pred))
print("DECISION TREE")
print("Accuracy score (validation): {0:.3f}".format(clf.score(x_test, y_test)))
print("Confusion Matrix Decision tree:")
print(confusion_matrix(y_test, y_pred_dt))
print()
print("Classification Report for Decision Tree:")
print(classification_report(y_test, y_pred_dt))
```
